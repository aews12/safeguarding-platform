```python
import pandas as pd

# Path to the CSV 
csv_path = r"C:\Users\Alex\Documents\dissertation-ai\safeguarding_platform\data\raw\safeguarding_reports_sample.csv"

# 1. Load the data
df = pd.read_csv(csv_path)

# 2. Show basic info
print("Number of reports:", len(df))
print("\nColumns in the dataset:")
print(df.columns.tolist())

print("\nPreview of key fields (first 5 rows):")
display(df[["report_id", "free_text", "category_label", "urgency_label"]].head())

# 3. Show how many examples per label (even though it's tiny right now)
print("\nCategory label counts:")
print(df["category_label"].value_counts())

print("\nUrgency label counts:")
print(df["urgency_label"].value_counts())
```

```python
# Define how we think about each column in the safeguarding dataset

USER_INPUT_FIELDS = [
    "timestamp",      # typically recorded at submission time
    "unit_id",
    "reporter_role",
    "location",
    "age_band",
    "channel",
    "free_text",
]

SYSTEM_FIELDS = [
    "report_id",      # system-generated ID
    "status",         # Open / In progress / Closed, etc.
]

LABEL_FIELDS = [
    "category_label", # ground truth concern type (for training/eval)
    "urgency_label",  # ground truth urgency level (for training/eval)
]

print("User input fields:")
for f in USER_INPUT_FIELDS:
    print(" -", f)

print("\nSystem fields:")
for f in SYSTEM_FIELDS:
    print(" -", f)

print("\nLabel fields (for model training/evaluation, NOT entered by user):")
for f in LABEL_FIELDS:
    print(" -", f)

# Quick sanity check: confirm all of these exist in the current DataFrame
print("\nSanity check against current df columns:")
missing = []
for f in USER_INPUT_FIELDS + SYSTEM_FIELDS + LABEL_FIELDS:
    exists = f in df.columns
    print(f"{f}: {'OK' if exists else 'MISSING'}")
    if not exists:
        missing.append(f)

if missing:
    print("\nWARNING: The following expected fields are missing from df:", missing)
else:
    print("\nAll expected fields are present in the dataset.")
```

```python
from transformers import pipeline

# Define the allowed urgency levels for classification
URGENCY_LABELS = ["Low", "Medium", "High", "Critical"]

# Create a zero-shot classification pipeline
# Note: First run will download the model, which can take a little while.
zero_shot_urgency_classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=0 if torch.cuda.is_available() else -1,
)

def classify_urgency_zero_shot(text: str) -> dict:
    """
    Classify the urgency of a safeguarding report using zero-shot learning.

    Returns a dictionary with:
      - 'predicted_urgency': the top predicted label
      - 'scores': all labels with their scores
    """
    result = zero_shot_urgency_classifier(
        text,
        candidate_labels=URGENCY_LABELS,
        multi_label=False,  # one best label
    )

    # The pipeline returns labels and scores; we bundle them nicely
    predicted = result["labels"][0]
    scores = dict(zip(result["labels"], result["scores"]))

    return {
        "predicted_urgency": predicted,
        "scores": scores,
    }

print("Zero-shot urgency classifier is ready.")
```

```python
# Let's test the zero-shot urgency classifier on one of our sample reports

test_row = df.iloc[3]  # this is the "bruises on arms" example (report_id 4)
print("Free text:")
print(test_row["free_text"])
print("\nGround truth urgency_label:", test_row["urgency_label"])

result = classify_urgency_zero_shot(test_row["free_text"])
print("\nZero-shot predicted urgency:", result["predicted_urgency"])
print("All scores:")
for label, score in result["scores"].items():
    print(f"  {label}: {score:.3f}")
```

```python
# Apply the zero-shot urgency classifier to all synthetic reports
# and compare with the ground truth urgency_label

results = []

for idx, row in df.iterrows():
    text = row["free_text"]
    true_urgency = row["urgency_label"]

    clf_result = classify_urgency_zero_shot(text)
    predicted_urgency = clf_result["predicted_urgency"]

    results.append({
        "report_id": row["report_id"],
        "free_text": text,
        "true_urgency": true_urgency,
        "predicted_urgency": predicted_urgency,
    })

comparison_df = pd.DataFrame(results)

print("Zero-shot urgency predictions vs ground truth:\n")
display(comparison_df[["report_id", "true_urgency", "predicted_urgency"]])

# Quick count of how many match / don't match
matches = (comparison_df["true_urgency"] == comparison_df["predicted_urgency"])
num_correct = matches.sum()
num_total = len(comparison_df)

print(f"\nCorrect predictions: {num_correct} / {num_total}")
print(f"Accuracy (on this tiny sample): {num_correct / num_total:.2f}")
```
